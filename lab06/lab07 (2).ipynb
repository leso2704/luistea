```python
#subiendo y graficando countries
```

```python
pais = pd.read_csv("countries.csv")
pais
```

```python
honduras = pais[pais.country == "Honduras"]
```

```python
ecuador = pais[pais.country == "Ecuador"]
```

```python
plt.plot(honduras.year, honduras.population/10**6)
plt.plot(ecuador.year, ecuador.population/10**6)
plt.legend(["Ecuador", "Honduras"])
plt.title("Poblacion Ecuador y Honduras")
plt.xlabel("Año")
plt.ylabel("Millones de Habitantes")
plt.show()
```
